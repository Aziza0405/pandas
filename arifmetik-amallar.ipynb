```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'e', 'g'])
s1
```

```python
s2 = pd.Series([6, 7, 8, 9, 10], index=['a', 'c', 'd', 'e', 'h'])
s2
```

```python
s1+s2     #Series'da ikkovida index bor bo'lishi kk
```

```python
d1 = pd.DataFrame(np.arange(9.).reshape(3, 3), columns=list('abc'), index=['Anor', 'Banan', 'Do\'lana'])
d1
```

```python
d2 = pd.DataFrame(np.arange(12.).reshape(4, 3), columns=list('bcd'), index=['Kivi', 'Anor', 'Banan', 'Nok'])
d2
```

```python
d1+d2           #DataFrame'da index va column ikkovida bor bo'lishi kk
```

```python
d1.add(d2, fill_value=1)        #fill value ikkovidan birida bor, index ham-column ham va value + x{fill_value}
```

**Boshqa arifmetik amallar**

add, radd - Qo'shish (+)

sub, rsub - Ayirish (-)

div, rdiv - Bo'lish (/)

floordiv, rfloordiv - Bo'lish va butun qismini olis (//)

mul, rmul - Ko'paytirish (*)

pow, rpow - Eksponenta (**)
